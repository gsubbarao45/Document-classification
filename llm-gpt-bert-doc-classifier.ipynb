```python
import os
import pandas as pd
import re

#!pip install nltk PyPDF2

import os
import pandas as pd
from PyPDF2 import PdfReader
import PyPDF2 


```

```python
# df = pd.DataFrame(columns=['File Name', 'Text'])
folder_path = r'pdf_data_unsup/'

data = []

for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'rb') as file:
            reader = PdfReader(file)
            text = ''
            for page in reader.pages:
                text += page.extract_text()
        data.append({'File Name': filename, 'Text': text})


df = pd.DataFrame(data)


print(df)
```

```python
df = df[["Text"]]
df.head()
```

```python
documents = df["Text"].to_list()
```

# 'gpt2'

```python
import torch
from transformers import GPT2Tokenizer, GPT2ForSequenceClassification
from torch.utils.data import DataLoader
from tqdm import tqdm

# Load pre-trained GPT-2 model and tokenizer
model_name = 'gpt2'  # or 'gpt2-medium', 'gpt2-large', 'gpt2-xl' for larger models
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token  # Set padding token to end-of-sequence token
model = GPT2ForSequenceClassification.from_pretrained(model_name)

# Tokenize and preprocess the documents
max_length = 128  # Maximum sequence length
batch_size = 1  # Set batch size to 1 to handle batch sizes > 1 without padding token
encoded_documents = tokenizer(documents, padding=True, truncation=True, max_length=max_length, return_tensors='pt')

# Create a DataLoader for the documents
dataset = torch.utils.data.TensorDataset(encoded_documents['input_ids'], encoded_documents['attention_mask'])
dataloader = DataLoader(dataset, batch_size=batch_size)

# Set the model to evaluation mode
model.eval()

# Classify the documents
predictions = []
with torch.no_grad():
    for batch in tqdm(dataloader, desc="Classifying"):
        input_ids, attention_mask = batch
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        probabilities = torch.softmax(logits, dim=1)
        predicted_labels = torch.argmax(probabilities, dim=1)
        predictions.extend(predicted_labels.cpu().numpy())

list(set(predictions))
```

```python
classified_docs = pd.DataFrame({'Document': documents, 'Predicted Label': predictions})


print(classified_docs)
```

# bert-base-uncased

```python


import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader
from tqdm import tqdm

# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'  # we can change the model as per our requirement
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)


# Tokenize and preprocess the documents
max_length = 128  # Maximum sequence length
batch_size = 5
encoded_documents = tokenizer(documents, padding=True, truncation=True, max_length=max_length, return_tensors='pt')

# Create a DataLoader for the documents
dataset = torch.utils.data.TensorDataset(encoded_documents['input_ids'], encoded_documents['attention_mask'])
dataloader = DataLoader(dataset, batch_size=batch_size)

# Set the model to evaluation mode
model.eval()

# Classify the documents
predictions = []
with torch.no_grad():
    for batch in tqdm(dataloader, desc="Classifying"):
        input_ids, attention_mask = batch
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        probabilities = torch.softmax(logits, dim=1)
        predicted_labels = torch.argmax(probabilities, dim=1)
        predictions.extend(predicted_labels.cpu().numpy())

predictions
```
